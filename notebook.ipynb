# Making an optimal marketing strategy for our insurance

We have been provided plenty of data from our customers and non customers from our agency. 

We will try using those data to find a way to plan a good marketing strategy to improve sales.

The main question we were asked is : 

<b>Are our customers more likely to travel abroad and travel more often compare to non customers ?  </b>

The data we were provided can help us make additional supposition to have a better insite on potential customers. 

Let's now have a quick look at the data we were provided : 




```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('travel_insurance.csv')
df.loc[df["TravelInsurance"]==1,'TravelInsurance']='Yes'
df.loc[df["TravelInsurance"]==0,'TravelInsurance']='No'
df.loc[df["ChronicDiseases"]==1,'ChronicDiseases']='Yes'
df.loc[df["ChronicDiseases"]==0,'ChronicDiseases']='No'
df.head()
```

Let's compare the fly habits of Customers vs Non customers :

```python
travel = pd.crosstab( df['FrequentFlyer'], df['TravelInsurance'])
print(travel)
fig, ax = plt.subplots(1,2) 
plt.title("Frequency Frequent Flyer/ Insurance")
ax[0].pie(travel['No'],colors=("r","g"), labels = ["No","Frequent"])
ax[0].set_xlabel('Without insurance')
ax[1].pie(travel['Yes'],colors=("r","g"),labels = ["No","Frequent"])
ax[1].set_xlabel('With insurance')
plt.show()
```

Let's see if Customers tends to fly abroad more :



```python
abroad = pd.crosstab( df['EverTravelledAbroad'], df['TravelInsurance'])
print(abroad)
fig, ax = plt.subplots(1,2) 
plt.title("Frequency Abroad Travel/ Insurance")
ax[0].pie(abroad['No'],colors=("r","g"), labels = ["Never","Yes"])
ax[0].set_xlabel('Without insurance')
ax[1].pie(abroad['Yes'],colors=("r","g"),labels = ["Never","Yes"])
ax[1].set_xlabel('With insurance')
plt.show()
```

Looking back on our data, we are going to make more insite :

Does the frenquency of flying is affected by annual income ? 

What about it when we consider customers ? 

```python
sns.catplot(x= "FrequentFlyer",y="AnnualIncome", hue = "TravelInsurance",data = df)
plt.title("Frequenct flyers vs Annual income")
plt.ylabel("Annual income in $ (100 K)")
plt.xlabel("Frequent Flyer")
plt.show()
```

```python
sns.boxplot(x= 'TravelInsurance', y= "AnnualIncome", data = df)
plt.title("Box plot of Annual Income VS Insurance")
plt.ylabel("Annual Income in $ (100K)")
plt.xlabel("Got travel insurance")
plt.show()
```









Seeing that we have data of family members, we can ask ourselves :

Does having more family members make you more likely to buy a travel insurance ? 












```python
sns.boxplot(x= 'TravelInsurance', y= "FamilyMembers", data = df) 
plt.title("Number of Family Members VS Travel Insurance") 
plt.xlabel("Got Travel Insurance")
plt.ylabel("Number of family members registered")
plt.show()
```

Does Age and graduation affect how likely they are to buy insurance ? 


```python
sns.swarmplot(x= 'TravelInsurance',y= "Age",hue="GraduateOrNot", data= df,linewidth=1,dodge=True)
plt.title("Age Vs Insurance with/without being graduated")
plt.yticks(range(25,40))
plt.show()


            
```

Which sector are more likely to travel and buy insurance ?

Lets look at our customers employment type and travel habits. 


```python
insurance = df[df["TravelInsurance"]=='Yes']
sns.countplot(x="Employment Type", hue= "FrequentFlyer",data = insurance)
plt.title("Number of client by sector and How many are frequent traveler")
plt.show()
```
